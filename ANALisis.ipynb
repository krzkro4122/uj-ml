```python
import pandas as pd

def load_dataset(filename: str = './dataset.csv') -> pd.DataFrame:
    return pd.read_csv(filename)

dataset_df = load_dataset()

print(dataset_df.to_string())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def sort_by_columns(df: pd.DataFrame):
    return df.set_index('Date').sort_values(by='May 2023', axis=1, ascending=False)

def plot(df: pd.DataFrame, **kwargs) -> None:
    
    df = sort_by_columns(df)
    df_plot = df.plot(figsize=(20, 10), rot=45,
        title="Most popular programming languages since 2004", 
        ylabel="Time since 2004",  
        xlabel="Popularity procent",  
        **kwargs)
    df_plot.set_xticks(range(len(df.index))[10::12], df.index[10::12])
    df_plot.set_yticklabels([f"{int(element)}%" for element in df_plot.get_yticks()])
    
    plt.legend(df.columns, loc='best', prop={'size': 12.5})
    plt.tight_layout()
    plt.show()

plot(dataset_df)
```
